Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10

Виведіть результат виконання кожної операції.

```python
arr = np.array([12, 24, 36, 48, 60])
print(f"Оригінальний масив: {arr}")
print(f"Результат множення на 2: {arr * 2}")
print(f"Результат додавання 10: {arr + 10}")
print(f"Результат піднесення до квадрата: {arr ** 2}")
print(f"Результат взяття квадратного кореня: {np.sqrt(arr)}")
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr = np.array([15, 22, 35, 44, 58, 67])
print(f"Дано масив: {arr}")
print(f"Перший елемент: {arr[0]}")
print(f"Останній елемент: {arr[-1]}")
print(f"Елементи з індексами 1 до 3 (включно): {arr[1:4]}")
print(f"Елементи більші за 40: {arr[arr>40]}")
```

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
```

```python
display(matrix)
print(f"Форма матриці: {matrix.shape}")
print(f"Перший рядок: {matrix[0,:]}")
print(f"Друга колонка: {matrix[:,1]}")
print(f"Суму всіх елементів: {np.sum(matrix)}")
print(f"Середнє всіх елементів: {np.mean(matrix)}")
```
