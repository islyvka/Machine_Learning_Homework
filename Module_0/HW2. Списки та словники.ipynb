**1. Списки - базові операції**

Створіть список температур за тиждень (у градусах Цельсія) з назвою
`temperatures` і значеннями [22, 25, 19, 23, 26, 20, 21, 22].


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть перший елемент
- Виведіть останній елемент
- Додайте температуру 24
- Видаліть температуру 19
- Виведіть довжину списку
- Обчисліть суму всіх температур
- Знайдіть максимальну температуру

```python
temperature = [22, 25, 19, 23, 26, 20, 21, 22]
```

```python
print(temperature[0]) # вивести перший елемент
print(temperature[-1]) # вивести останній елемент
temperature.append(24) # додати 24
temperature.remove(19) # видалити 19
print(len(temperature)) # вивести довжину списку
print(sum(temperature)) # обчислити та вивести суму
print(max(temperature)) # знайти та вивести максимальну температуру
```


**2. Словники - базові операції**

Створіть словник про себе, який містить ключі `name`, `age`, `city` і значення, які відповідають вашему імені, віку і місту (можна вставити будь-які значення, не обовʼязково свої).


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть значення ключа "name"
- Змініть значення ключа "age" на 21
- Додайте новий ключ "hobby" зі значенням "programming"
- Виведіть всі ключі
- Виведіть всі значення


```python
customer = {
    'name': 'Ivanna',
    'age': 34,
    'city': 'Kyiv'
}
```

```python
print(customer['name']) #вивести значення ключа "name"
```

```python
customer['age'] = 21 # замінити значення ключа "age" на 21
print(customer)
```

```python
customer['hobby'] = 'programming' # додати новий ключ "hobby" зі значенням "programming"
print(customer)
```

```python
print(customer.keys()) # вивести всі ключі
```

```python
print(customer.values()) # вивести всі значення
```
