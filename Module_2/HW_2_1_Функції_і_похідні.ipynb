1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# задаємо функцію
def f(x):
    return 10*x-5

# її похідну
def derivative(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

# точки
x = np.linspace(0, 10, 100)

# обчислюємо похідну
df = derivative(f, x)

# наносимо на графік обидві функції
plt.figure(figsize=(10, 5))
plt.plot(x, f(x), label="f(x) = 10x - 5")
plt.plot(x, df, label="f'(x) = 10")

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

**Характер залежності:** Функція $f(x) = 10x - 5$ є лінійною. Це означає, що кожна додаткова година навчання приносить фіксовану кількість балів.

**Швидкість зростання:** Обчислена похідна дорівнює 10. Це стале число, яке показує, що швидкість отримання знань не змінюється з часом. Учень стабільно отримує +10 балів за кожну витрачену годину.

**Вплив на оцінку:** Зі збільшенням кількості годин оцінка зростає рівномірно.Точка перетину з віссю $x$ ($10x - 5 = 0$) — це 0.5 години. Тобто потрібно витратити часу строго більше за пів години, щоб почати отримувати позитивний результат (оцінку вище нуля). При максимально заданих 10 годинах навчання прогнозована оцінка становитиме: $10 \times 10 - 5 = 95$ балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Функція синус
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)

# Логарифмічна функція
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна функції sin(x):", derivative_f_sin)
print("Похідна логарифмічної функції ln(x):", derivative_f_ln)
```

```python
# Графік функції sin(x) та її похідної
xs = np.linspace(-10, 10, 100)

plt.figure(figsize=(10, 5))
plt.plot(xs, np.sin(xs), label='f(x) = sin(x)')
plt.plot(xs, np.cos(xs), label="f'(x) = cos(x)")
plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Графік функції ln(x) та її похідної
xs = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 5))
plt.plot(xs, np.log(xs), label='f(x) = ln(x)')
plt.plot(xs, 1/xs, label="f'(x) = 1/x")
plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# символ
x = sp.symbols('x')

# функція
f_sym = x**2 - 5*x + 1
df_sym = sp.diff(f_sym, x)

print(f"Похідна функції f(x): {df_sym}")

# числові функції
f = sp.lambdify(x, f_sym, 'numpy')
df = sp.lambdify(x, df_sym, 'numpy')

# діапазон
x_vals = np.linspace(0, 100, 1000)

# графік
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f(x_vals), label=f"f(x) = {f_sym}")
plt.plot(x_vals, df(x_vals), label=f"f'(x) = {df_sym}")

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
